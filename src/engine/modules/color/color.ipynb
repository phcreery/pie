```python

# cam_mul: 1.90625, 1, 1.4921875, 1
# pre_mul: 2.2319138, 0.93263817, 1.1414729, 0

# xyz_d65_from_cam:
# 1.5699793 0.14481506 0.10475578
# 0.6360732 0.8477491 -0.221411
# 0.12701747 -0.2259345 1.4544346


# multiply wb by matrix
np_wb_cam_mul = np.array([1.90625, 1, 1.4921875, 1], dtype=np.float32)
# np_wb_pre_mul = np.array([2.2319138, 0.93263817, 1.1414729, 0], dtype=np.float32)
np_xyz_d65_from_cam = np.array([[1.5699793, 0.14481506, 0.10475578],
                             [0.6360732, 0.8477491, -0.221411],
                             [0.12701747, -0.2259345, 1.4544346]], dtype=np.float32)

np_srgb_from_xyz = np.array([[3.2404542, -1.5371385, -0.4985314],
                             [-0.969266, 1.8760108, 0.041556],
                             [0.0556434, -0.2040259, 1.0572252]], dtype=np.float32)

np_rec2020_from_xyz = np.array([[1.7166512, -0.3556708, -0.2533663],
                                [-0.6666844, 1.6164812, 0.0157685],
                                [0.0176399, -0.0427706, 0.9421031]], dtype=np.float32)

print("WB cam mul:", np_wb_cam_mul)
# print("WB pre mul:", np_wb_pre_mul)

result_xyz = np.dot(np_xyz_d65_from_cam, 1/np_wb_cam_mul[:3])
# result_xyz = np.dot(np_xyz_d65_from_cam, 1/np_wb_pre_mul[:3])
# result_xyz /= result_xyz[1]
print("Result XYZ:", result_xyz)

result_srgb = np.dot(np_srgb_from_xyz, result_xyz)
result_srgb /= result_srgb[1]
print("Result sRGB:", result_srgb)
result_srgb_pwb = 1/result_srgb[:3]
print("Result sRGB (pre-WB):", result_srgb_pwb) # this is how its done in vkdt

result_rec2020 = np.dot(np_rec2020_from_xyz, result_xyz)
result_rec2020 /= result_rec2020[1]
print("Result Rec.2020:", result_rec2020)
result_rec2020_pwb = 1/result_rec2020[:3]
print("Result Rec.2020 (pre-WB):", result_rec2020_pwb) # this is how its done in vkdt
```
